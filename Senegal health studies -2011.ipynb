# Health Data Analysis in Senegal

This Jupyter Notebook performs an analysis of health indicators in Senegal, focusing on open defecation, access to drinking water, access to basic sanitation services, and child health indicators. The analysis includes both descriptive statistics and comparative analysis between different years and quintiles.


## Data Description

The dataset used in this analysis contains health and nutrition data for Senegal, including information on open defecation, access to clean water, access to sanitation services, and child health indicators. The dataset covers multiple years and quintiles.


## Descriptive Analysis

### Open Defecation

We analyze the percentage of the population practicing open defecation in Senegal, broken down by quintiles and across different years (2011 and 2015).

### Access to Drinking Water

We examine the percentage of the population with access to basic drinking water services, both for rural and urban areas. The analysis covers multiple quintiles and includes data for 2011 and 2015.

### Access to Basic Sanitation Services

We investigate the percentage of the population with access to basic sanitation services in Senegal. The analysis considers rural and urban areas, various quintiles, and data from 2011 and 2015.

### Child Health Indicators

We assess various child health indicators, including the prevalence of acute respiratory infection (ARI), diarrhea prevalence, diarrhea treatment, and under-five mortality rate. The analysis involves comparing these indicators across quintiles for the years 2011 and 2015.


## Comparative Analysis

### Open Defecation

In 2015, we observe a significant decrease in the percentage of the population practicing open defecation across all quintiles compared to 2011. Quintile 5 (highest) consistently has the lowest open defecation rates.

### Access to Drinking Water

Access to basic drinking water services has improved from 2011 to 2015 for both rural and urban populations. Quintile 1 (lowest) experiences the most substantial improvements.

### Access to Basic Sanitation Services

Senegal has made progress in providing basic sanitation services, with notable improvements in urban areas. However, rural areas, especially in lower quintiles, still lag behind in access to sanitation services.

### Child Health Indicators

The prevalence of acute respiratory infection (ARI) among children under 5 has decreased in 2015 compared to 2011. A similar trend is observed for diarrhea prevalence. However, ARI treatment rates remain low in the highest quintile.

The under-five mortality rate has seen an overall decrease, reflecting improvements in child health in Senegal.
